La **distribución binomial** modela el número de éxitos en una serie de experimentos simultáneos e independientes, donde la probabilidad de éxito es fijada de antemano. 

En los casinos ingleses del siglo XVII se practicaba un juego en el que se arrojaban 9 monedas al aire. Si se obtenían menos de 5 caras, entonces se perdía una libra, si no se ganaba una libra. Con Numpy se puede simular "la suerte" de un apostador empedernido que va una noche al casino con 1000 libras para apostar.

Primero se crea un vector, utilizando `numpy.zeros`, que representará la evolución del capital del jugador durante la partida.

```python
import numpy
credito = numpy.zeros(10000)   #Juega 10000 veces, por eso es un jugador empedernido :)
credito[0] = 1000
```

Después con la función `numpy.random.binomial(N° experimentos simultáneos, prob. de éxito, N° jugadas)` se simulan los resultados de las jugadas.

```python
suerte = np.random.binomial(9, 0.5, size=credito.size)
```

¿Cómo le fue? Actualizamos el vector de crédito después de cada jugada.

```python
for i in xrange(1, len(credito)):
    if suerte[i] < 5:
        credito[i] = credito[i - 1] - 1
    else:
        credito[i] = credito[i - 1] + 1
```

Finalmente pueden graficarse los resultados.

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot(credito)
plt.title('Suerte del jugador')
plt.xlabel('Numero de jugada')
plt.ylabel('Capital (Libras)')
plt.show()
```
